```python
def regression(x, y, color):
    reg = linregress(x, y)
    corr, _ = pearsonr(x, y)
    if corr > 0:
        print('Positive correlation')
    elif corr < 0:
        print('Negative correlation')
    else:
        print('No correlation')
    print('Pearsons correlation: %.3f' % corr)
    print(reg)
    sns.regplot(x, y, color=color)
```

```python
#  'pct_hh_rent_burd'   'unemployment_cd'     'pct_hispanic'
regression(big_data['pct_hispanic'], big_data['unemployment_cd'], 'green')
```
